```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

%matplotlib inline
sns.set_style("whitegrid")
```

```python
from google.colab import files
uploaded = files.upload()

df = pd.read_csv("Life Expectancy Data.csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
# Rename columns (remove spaces for easier coding)
df.columns = df.columns.str.replace(' ', '_')

# Drop rows with too many missing values OR fill them
df = df.dropna()

# Check again
df.isnull().sum()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df.Life_expectancy_, bins=30, kde=True)
plt.title("Distribution of Life Expectancy")
plt.xlabel("Life Expectancy")
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Status', y='Life_expectancy_')
plt.title("Life Expectancy: Developed vs Developing Countries")
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include='number')

# Drop rows with missing values (if any remain)
numeric_df = numeric_df.dropna()

# Plot correlation heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation between Health Features")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='GDP', y='Life_expectancy_', hue='Status')
plt.title("GDP vs Life Expectancy")
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Schooling', y='Life_expectancy_', hue='Status')
plt.title("Schooling vs Life Expectancy")
```

```python
fig = px.line(df[df['Country'] == 'India'], x='Year', y='Life_expectancy_', title='India - Life Expectancy Over Time')
fig.show()
```
